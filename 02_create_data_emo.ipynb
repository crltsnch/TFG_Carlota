```python
import pandas as pd

# 1. Cargar el CSV con el riesgo cl√≠nico
df = pd.read_csv("Dataframe/data_riesgo_clinico.csv")

# Filtrar y contar pacientes con riesgo > 70
n_mayores_70 = (df["riesgo_clinico"] > 70).sum()

print(f"üìä Pacientes con riesgo cl√≠nico > 70: {n_mayores_70}")
```

```python
import numpy as np
q_low = np.percentile(df["riesgo_clinico"], 33)
q_high = np.percentile(df["riesgo_clinico"], 52)
```

```python
import numpy as np

def generar_variables_emocionales(riesgo_clinico, q_low, q_high):
    if riesgo_clinico > q_high:
        responsabilidad = np.random.randint(1, 2)
        neuroticismo = np.random.randint(4, 6)
        extraversion = np.random.randint(1, 2)
        amabilidad = np.random.randint(1, 2)
        apertura = np.random.randint(1, 2)

        ansiedad = np.random.choice([0, 1, 2, 3], p=[0.1, 0.1, 0.4, 0.4])
        depresion = np.random.choice([0, 1, 2, 3], p=[0.1, 0.1, 0.4, 0.4])
        estres = np.random.choice([0, 1, 2, 3], p=[0.1, 0.1, 0.4, 0.4])

        apoyo = np.random.choice(["No", "S√≠"], p=[0.8, 0.2])

        alcohol = np.random.choice(["Nunca", "Ocasionalmente", "Frecuentemente"], p=[0.1, 0.4, 0.5])
        fuma_actual = np.random.choice(["No", "S√≠, <5 cigarros/d√≠a", "S√≠, >5 cigarros/d√≠a"], p=[0.1, 0.2, 0.7])
        fue_fumador = np.random.choice(["No", "S√≠, en el pasado", "S√≠, sigue si√©ndolo"], p=[0.1, 0.3, 0.6])
        alcohol_problema = np.random.choice(["No", "S√≠, en el pasado", "S√≠, sigue teniendo actualmente"], p=[0.1, 0.3, 0.6])
        drogas_pasado = np.random.choice(["No", "Ocasionalmente", "Frecuentemente"], p=[0.1, 0.4, 0.5])
        drogas_ahora = np.random.choice(["No", "Ocasionalmente", "Habitualmente"], p=[0.2, 0.5, 0.3])
        drogas_problema = np.random.choice(["No", "S√≠"], p=[0.2, 0.8])

        ejercicio = np.random.choice(["No", "S√≠"], p=[0.8, 0.2])
        alimentacion = np.random.choice(["No", "S√≠"], p=[0.8, 0.2])

    elif riesgo_clinico < q_low:
        responsabilidad = np.random.randint(4, 6)
        neuroticismo = np.random.randint(1, 2)
        extraversion = np.random.randint(4, 6)
        amabilidad = np.random.randint(4, 6)
        apertura = np.random.randint(4, 6)

        ansiedad = np.random.choice([0, 1, 2, 3], p=[0.5, 0.3, 0.1, 0.1])
        depresion = np.random.choice([0, 1, 2, 3], p=[0.5, 0.3, 0.1, 0.1])
        estres = np.random.choice([0, 1, 2, 3], p=[0.5, 0.3, 0.1, 0.1])

        apoyo = np.random.choice(["No", "S√≠"], p=[0.2, 0.8])  #0 no tiene apoyo, 1 tiene apoyo

        alcohol = np.random.choice(["Nunca", "Ocasionalmente", "Frecuentemente"], p=[0.7, 0.25, 0.05])
        fuma_actual = np.random.choice(["No", "S√≠, <5 cigarros/d√≠a", "S√≠, >5 cigarros/d√≠a"], p=[0.8, 0.15, 0.05])
        fue_fumador = np.random.choice(["No", "S√≠, en el pasado", "S√≠, sigue si√©ndolo"], p=[0.7, 0.2, 0.1])
        alcohol_problema = np.random.choice(["No", "S√≠, en el pasado", "S√≠, sigue teniendo actualmente"], p=[0.9, 0.05, 0.05])
        drogas_pasado = np.random.choice(["No", "Ocasionalmente", "Frecuentemente"], p=[0.85, 0.1, 0.05])
        drogas_ahora = np.random.choice(["No", "Ocasionalmente", "Habitualmente"], p=[0.9, 0.08, 0.02])
        drogas_problema = np.random.choice(["No", "S√≠"], p=[0.7, 0.3])

        ejercicio = np.random.choice(["No", "S√≠"], p=[0.2, 0.8])
        alimentacion = np.random.choice(["No", "S√≠"], p=[0.2, 0.8])

    else:
        responsabilidad = np.random.randint(2, 5)
        neuroticismo = np.random.randint(2, 5)
        extraversion = np.random.randint(2, 5)
        amabilidad = np.random.randint(2, 5)
        apertura = np.random.randint(2, 5)

        ansiedad = np.random.choice([0, 1, 2, 3], p=[0.2, 0.3, 0.4, 0.1])
        depresion = np.random.choice([0, 1, 2, 3], p=[0.2, 0.3, 0.4, 0.1])
        estres = np.random.choice([0, 1, 2, 3], p=[0.2, 0.3, 0.4, 0.1])

        apoyo = np.random.choice(["No", "S√≠"])

        alcohol = np.random.choice(["Nunca", "Ocasionalmente", "Frecuentemente"], p=[0.4, 0.5, 0.1])
        fuma_actual = np.random.choice(["No", "S√≠, <5 cigarros/d√≠a", "S√≠, >5 cigarros/d√≠a"], p=[0.4, 0.5, 0.1])
        fue_fumador = np.random.choice(["No", "S√≠, en el pasado", "S√≠, sigue si√©ndolo"], p=[0.4, 0.5, 0.1])
        alcohol_problema = np.random.choice(["No", "S√≠, en el pasado", "S√≠, sigue teniendo actualmente"], p=[0.4, 0.5, 0.1])
        drogas_pasado = np.random.choice(["No", "Ocasionalmente", "Frecuentemente"], p=[0.4, 0.5, 0.1])
        drogas_ahora = np.random.choice(["No", "Ocasionalmente", "Habitualmente"], p=[0.4, 0.5, 0.1])
        drogas_problema = np.random.choice(["No", "S√≠"], p=[0.4, 0.6])

        ejercicio = np.random.choice(["No", "S√≠"])
        alimentacion = np.random.choice(["No", "S√≠"])

    return {
        "responsabilidad": responsabilidad,
        "neuroticismo": neuroticismo,
        "extraversi√≥n": extraversion,
        "amabilidad": amabilidad,
        "apertura": apertura,
        "ansiedad": ansiedad,
        "depresion": depresion,
        "estres": estres,
        "apoyo": apoyo,
        "alcohol": alcohol,
        "fuma_actual": fuma_actual,
        "fue_fumador": fue_fumador,
        "alcohol_problema": alcohol_problema,
        "drogas_pasado": drogas_pasado,
        "drogas_ahora": drogas_ahora,
        "drogas_problema": drogas_problema,
        "ejercicio": ejercicio,
        "alimentacion": alimentacion,
    }
```

```python
variables_emocionales = df["riesgo_clinico"].apply(
    lambda r: generar_variables_emocionales(r, q_low, q_high)
)

# Convertir a DataFrame y concatenar
df_emocionales = pd.DataFrame(variables_emocionales.tolist())
df_completo = pd.concat([df, df_emocionales], axis=1)
```

```python
from sklearn.preprocessing import OrdinalEncoder
import joblib
import os

# Crear carpeta para guardar los encoders
os.makedirs("Outputs/encoder_emo", exist_ok=True)

# Definir manualmente los √≥rdenes
ordenes_personalizados = {
    "apoyo": ["No", "S√≠"],
    "alcohol": ["Nunca", "Ocasionalmente", "Frecuentemente"],
    "fuma_actual": ["No", "S√≠, <5 cigarros/d√≠a", "S√≠, >5 cigarros/d√≠a"],
    "fue_fumador": ["No", "S√≠, en el pasado", "S√≠, sigue si√©ndolo"],
    "alcohol_problema": ["No", "S√≠, en el pasado", "S√≠, sigue teniendo actualmente"],
    "drogas_pasado": ["No", "Ocasionalmente", "Frecuentemente"],
    "drogas_ahora": ["No", "Ocasionalmente", "Habitualmente"],
    "drogas_problema": ["No", "S√≠"],
    "ejercicio": ["No", "S√≠"],
    "alimentacion": ["No", "S√≠"]
}

codificaciones = {}

# Entrenar y guardar cada encoder manualmente
for col, orden in ordenes_personalizados.items():
    encoder = OrdinalEncoder(categories=[orden])
    encoder.fit([[val] for val in orden])  # ENTRENAMIENTO MANUAL, seguro y estable
    joblib.dump(encoder, f"Outputs/encoder_emo/encoder_emo_{col}.pkl")
    codificaciones[col] = dict(zip(orden, range(len(orden))))
    print(f"Codificaci√≥n para '{col}': {codificaciones[col]}")
```
