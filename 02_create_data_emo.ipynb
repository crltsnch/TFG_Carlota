```python
variables_emocionales = df["riesgo_clinico"].apply(
    lambda r: generar_variables_emocionales(r, q_low, q_high)
)

# Convertir a DataFrame y concatenar
df_emocionales = pd.DataFrame(variables_emocionales.tolist())
df_completo = pd.concat([df, df_emocionales], axis=1)
```

```python
from sklearn.preprocessing import LabelEncoder
import joblib
import os

# Crear el directorio si no existe
os.makedirs("../Outputs/encoder", exist_ok=True)

# Seleccionar las columnas categÃ³ricas
categorical_cols = df_completo.select_dtypes(include=['object', 'category']).columns.tolist()

# Crear el LabelEncoder
le = LabelEncoder()

# Transformar las columnas categÃ³ricas y mostrar el mapeo
codificaciones = {}  # Diccionario para almacenar las codificaciones

for col in categorical_cols:
    le.fit(df_completo[col])  # Ajustar el LabelEncoder a la columna
    df_completo[col] = le.transform(df_completo[col])  # Transformar la columna
    joblib.dump(le, f"Outputs/encoder_emo/encoder_emo_{col}.pkl")  # Guardar codificador
    
    # Guardar el mapeo de valores originales y codificados
    codificaciones[col] = dict(zip(le.classes_, le.transform(le.classes_)))

# Imprimir las codificaciones
for col, mapping in codificaciones.items():
    print(f"CodificaciÃ³n para '{col}': {mapping}")
```
