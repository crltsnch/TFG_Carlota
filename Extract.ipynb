```python
import xenaPython as xena
import pandas as pd
```

```python
# Hub y dataset
hub = "https://gdc.xenahubs.net"
dataset = "TCGA-PRAD.clinical.tsv"

# Obtener todas las muestras
samples = xena.dataset_samples(hub, dataset, None)

# Obtener todas las variables disponibles correctamente
variables = xena.dataset_field(hub, dataset)

# Descargar todos los valores para todas las variables
position, values = xena.dataset_probe_values(hub, dataset, samples, variables)

# Crear DataFrame
df = pd.DataFrame({var: values[i] for i, var in enumerate(variables)})
df.insert(0, "sample_id", samples)

# Guardar como CSV
df.to_csv("TCGA_PRAD_clinical_full.csv", index=False)
print("✅ Archivo guardado como 'TCGA_PRAD_clinical_full.csv'")
```
